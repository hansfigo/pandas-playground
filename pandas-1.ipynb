## Filter Specific Row from DataFrame
to filter row based a certain condition we use condition inside `[]`

```python
# Show only fruit that fresh
fresh_fruit = df[df["Form"] == "Fresh"]
fresh_fruit.head()
```

```python
# Show only fruit that price is lower than 3
cheap = df[df["RetailPrice"] < 3]
cheap.head()
```

### isin ()
it will filter a row based on list of value 

```python
form_fresh_juice = df[df["Form"].isin(["Fresh", "Juice"])]
form_fresh_juice.head()
```

## Select Specific Row and Column in DataFrame
suppose we wanna know the name of fruits that have retail price more than 4

we use `loc` when we wanna select row and column using row and column name

```python
expensive_fruits = df.loc[df["RetailPrice"] >= 4 , ["Fruit", "RetailPrice"]]
expensive_fruits
```

### suppose we wanna show row 12 until 54 and column 2 to 4
